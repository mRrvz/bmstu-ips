```python
from pymystem3 import Mystem
from razdel import tokenize

mystem = Mystem()

mystem_result = mystem.lemmatize(text)
razdel_result = tokenize(text)
razdel_result = list(map(lambda it : it.text, list(razdel_result)))
```

```python
print(len(mystem_result))
print(mystem_result)
print(len(razdel_result))
print(razdel_result)

mystem_result_fixed = list()
for item in mystem_result:
  if item not in (' ', '\n'):
    mystem_result_fixed.append(item)

print(len(mystem_result_fixed))
for i1, i2 in zip(mystem_result_fixed, razdel_result):
  if i1 != i2:
    print(f"mystem: {i1}\nrazdel: {i2}\n")
```

## 3. Лемматизация Mystem vs Pymorphy
Лемматизируйте текст с помощью mystem и pymorphy. Найдите различия в лемматизации. Что, по вашему мнению, работает лучше на этом тексте?

```python
!pip install pymorphy3
from pymorphy3 import MorphAnalyzer
import nltk

nltk.download("punkt")

morph = MorphAnalyzer()

mystem_result = mystem.lemmatize(text)
pymorphy_result = [morph.parse(word)[0].normal_form for word in nltk.word_tokenize(text)]
pymorphy_result = list(map(lambda it: it if it == ',' or it == '.' else " " + it, pymorphy_result))

print(''.join(mystem_result))
print("=====")
print("".join(pymorphy_result))
```
